## Multiple plots

```python
fig, a = plt.subplots(2,2)     #a is an array can be scalar vector tensor anything
x = np.arange(1,15)

a[0][0].plot(x,np.sin(x**2), marker='o')
a[0][1].plot(x,np.sqrt(x), marker='x')
a[1][0].plot(x,np.exp(x), marker='p')
a[1][1].plot(x,np.log10(x), marker='s')
fig.suptitle('Math Sub plots')
a[0][0].set_title('Squared')
a[0][1].set_title('Square Root')
a[1][0].set_title('Exponential')
a[1][1].set_title('Log 10')

#fig.tight_layout()
fig.tight_layout(pad=2.5)

plt.show()
```

## Bar charts



```python
favorites = {"Almond Butter": 10,
            "Blueberries": 5,
            "Eggs": 6,
            "Cup cakes": 8}

fig, ax = plt.subplots()

ax.bar(favorites.keys(), 
       favorites.values(), 
       color = "Purple", 
       width = .3)

plt.show()
```

## Histogram

```python
marks = [30,60,78,90,31,57,80,99,38,94,78,89,62]

grades = [0,35,65,100]      #intervals

plt.hist(marks, bins = grades, color = "Gray")
plt.xticks([0,35,65,100] )
plt.show()
```

## Stacked bar chart

```python
years = ['2019','2020','2021','2022']
placements = [200,250, 300, 290]
```

```python
cse = [70, 90, 104, 70]
it = [22,50, 0, 104]
ece = [108, 110, 106, 116]
```

```python
ece_start = [cse[i] + it[i]  for i in range(len(cse))]
```

```python
plt.bar(years, cse, width = .4, color = 'Black', label = 'CSE')
plt.bar(years, it, bottom=cse, width = .4, color = 'Grey', label = 'IT')
plt.bar(years, ece, bottom=ece_start, width = .4, color = 'Silver', label = 'ESE')

plt.ylim(0,max(placements)+50)
plt.legend()
plt.show()
```

## Pie chart

```python
y = np.array([35,25, 15,17,21])
myLabels = ["Blueberry","Gooseberry","Raspberry", "Blackberry", "Strawberry"]
plt.figure(figsize=(6,4))
plt.pie(y,labels = myLabels)
plt.legend(bbox_to_anchor = (2,1))
plt.show()
```
